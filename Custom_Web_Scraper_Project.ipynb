```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"

# Fetch the HTML content
response = requests.get(url)
response.raise_for_status()

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table element (adjust selector if needed)
table = soup.find('table')

# Extract table data into a DataFrame and skip the first
# row to avoid header duplication
df = pd.read_html(str(table), skiprows=0)[0]

# Drop this columns
df.drop(columns=["State-owned", "Ref."], inplace=True)

# Rename Headquarters Column name
df.rename(columns={
    "Headquarters[note 1]": "Headquarters"
    }, inplace=True)

# Explicitly set the column names
df.columns = ['Rank', 'Name', 'Industry', 'Revenue USD millions', 'Profit USD millions',
       'Employees', 'Headquarters', 'Revenue per worker']
# Save the DataFrame to a CSV file
df.to_csv("largest_companies_revenue.csv", index=False)

print("CSV file created successfully!")
```
